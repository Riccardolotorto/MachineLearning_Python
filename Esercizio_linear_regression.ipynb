```python
# Import the required libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Read the CSV file:
data = pd.read_csv("/content/drive/MyDrive/Machine Learning Python/Machine Learning/data/energydata_complete.csv")
data.head()
```

```python
data['consumoTotale'] = data['rv1'] + data['rv2']
data.head()
```

```python
data = data.drop(['rv1', 'rv2', 'Tdewpoint'], axis = 1)
data.head()
```

```python
# Consider features we want to work on:
X = data[[ 'T1', 'RH_1', 'T2','RH_2', 'T3','RH_3', 'T4', 'RH_4', 'T5', 'RH_5', 'T6', 'RH_6', 'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'Press_mm_hg', 'RH_out', 'Windspeed']]
X
```

```python
Y = data['consumoTotale']
Y
```

```python
Xnp = X.to_numpy()
Xnp
```

```python
Ynp = Y.to_numpy()
Ynp
```

```python
from sklearn.model_selection import train_test_split # suddivide le osservazioni nei due insiemi

X_train, X_test, y_train, y_test = train_test_split(Xnp, Ynp,
                                                    test_size = 0.33,
                                                    random_state = 123) # fix random seed for replicability
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
print(X_train[:10])
print(f'y_train = {y_train[:10]}')
```

```python
#Using sklearn package to model data :
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

```python
predict = regr.predict(X_test)
```

```python
print(f'dati predetti dal metodo = {predict[:10].round(decimals=2)}') # dati predetti dal metodo
print(f'dati reali presenti nel dataset di test = {y_test[:10]}') # dati effettivi presenti nel test set
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# The coefficients
print('Coefficients: \n', regr.coef_)

# The mean squared error
print('Mean squared error: %.2f' %mean_squared_error(y_test, predict))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f' %r2_score(y_test, predict))
```
